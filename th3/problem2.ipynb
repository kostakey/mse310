```python
import mpmath as mp

R = 8.314
T = 1100.0

# Free energies (example forms from your statement)
def DG1(T):
    return 220000.0 - 210.0*T

def DG2(T):
    return -28000.0 + 32.0*T

K1 = mp.e**(-DG1(T)/(R*T))
K2 = mp.e**(-DG2(T)/(R*T))

def eq1(xi1, xi2):
    n_CH4 = 1.0 - xi1
    n_H2O = 1.0 - xi1 - xi2
    n_CO  = xi1 - xi2
    n_CO2 = xi2
    n_H2  = 3.0*xi1 + xi2

    n_tot = n_CH4 + n_H2O + n_CO + n_CO2 + n_H2

    y_CH4 = n_CH4/n_tot
    y_H2O = n_H2O/n_tot
    y_CO  = n_CO/n_tot
    y_CO2 = n_CO2/n_tot
    y_H2  = n_H2/n_tot

    return (y_CO * y_H2**3)/(y_CH4*y_H2O) - K1

def eq2(xi1, xi2):
    n_CH4 = 1.0 - xi1
    n_H2O = 1.0 - xi1 - xi2
    n_CO  = xi1 - xi2
    n_CO2 = xi2
    n_H2  = 3.0*xi1 + xi2

    n_tot = n_CH4 + n_H2O + n_CO + n_CO2 + n_H2

    y_CH4 = n_CH4/n_tot
    y_H2O = n_H2O/n_tot
    y_CO  = n_CO /n_tot
    y_CO2 = n_CO2/n_tot
    y_H2  = n_H2 /n_tot

    return (y_CO2*y_H2)/(y_CO*y_H2O) - K2

# Solve
xi1_sol, xi2_sol = mp.findroot((eq1, eq2), (0.40, 0.10))

print("xi1 =", xi1_sol)
print("xi2 =", xi2_sol)
```
