```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

n = 1.0                          # moles
R = 0.082057                     # L·atm/(mol·K)
T_values = np.arange(300, 1301, 100)  # Temperatures: 300 to 1300 K
volumes = [11.2, 22.4, 44.8]     # L
```

```python
# Ideal gas law: p = nRT / V
data = {'T (K)': T_values}
for V in volumes:
    p_values = (n * R * T_values) / V
    # rounded to 3 sig figs
    p_values = [float(f"{p:.3g}") for p in p_values]
    data[f'p (atm) at V = {V:.3f} L'] = p_values

df = pd.DataFrame(data)
```

```python
# Display table
print("Pressure vs. Temperature for 1 mol Ideal Gas")
print(df.to_string(index=False))
```

```python
# Display slopes
print("\nSlopes (dp/dT = nR/V):")
for V in volumes:
    slope = (n * R) / V
    print(f"  V = {V:>5.1f} L  -->  dp/dT = {slope:.3g} atm/K")
```

```python
# Pressure vs. temperature
plt.figure(figsize=(8,5))
for V in volumes:
    pressures = (n*R*T_values)/V
    plt.plot(T_values, pressures, marker='o', label=f'V = {V} L')

plt.title('p vs T (1 mol ideal gas) for fixed volumes')
plt.xlabel('Temperature T (K)')
plt.ylabel('Pressure p (atm)')
plt.grid(True)
plt.legend()
# plt.tight_layout()
plt.show()
```
